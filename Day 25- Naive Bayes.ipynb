# Naive Bayes Algorithm

### Importing packages

```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import *
from sklearn.metrics import accuracy_score , confusion_matrix
```

### Loading the Dataset

```python
data = pd.read_csv("train.csv")
```

```python
data.head()
```

### Converting text to num

```python
le = preprocessing.LabelEncoder()
data["Sex"]= le.fit_transform(data["Sex"])
data["Embarked"]= le.fit_transform(data["Embarked"])
data.head()
```

### Building the Model

```python
data.columns
```

```python
data = data.drop(['PassengerId', 'Name', 'Age',  'Ticket', 'Fare', 'Cabin',],axis=1)
```

```python
for i in data:
    y = data[i]
    X = data.drop([i],axis=1)
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
    clf = BernoulliNB()
    y_pred = clf.fit(X_train,y_train).predict(X_test)
    print("Accuracy score when "+i+" is Dependent variable: " ,accuracy_score(y_test,y_pred, normalize=True))
    print("Confusion matrix when "+i+" is Dependent variable:\n ",confusion_matrix(y_test,y_pred))
    print("\n")
    
```
